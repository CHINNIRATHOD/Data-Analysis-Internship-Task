```python
import pandas as pd

df = pd.read_csv("../data/ipl.csv")

df["runs_total"] = pd.to_numeric(df["runs_total"], errors="coerce")
df["over"] = pd.to_numeric(df["over"], errors="coerce")
df["ball"] = pd.to_numeric(df["ball"], errors="coerce")
```

```python
df["current_score"] = df.groupby(["match_id", "batting_team"])["runs_total"].cumsum()
```

```python
df["is_wicket"] = df["wicket_kind"].notnull().astype(int)

df["wickets_fallen"] = df.groupby(["match_id", "batting_team"])["is_wicket"].cumsum()
```

```python
df["runs_last_5"] = df.groupby(["match_id", "batting_team"])["runs_total"] \
    .rolling(window=30, min_periods=1).sum().reset_index(0, drop=True)
```
